```python
import pandas as pd
import numpy as np
df = pd.read_csv("health.csv")
# Menghilangkan kolom 'Unnamed: 0' jika ada
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

for column in df.columns:
    col = pd.Series(df[column])
    dtype = df[column].dtype
    print(f"Descriptive Statistics for column: {column}")
    
    if dtype != "object":
        # Data numerik
        print("Mean: ", round(col.mean(), 3))
        print("Median: ", round(col.median(), 3))
        print("Mode: ", [round(val, 3) for val in col.mode().values])
        print("Standard Deviation: ", round(col.std(), 3))
        print("Variance: ", round(col.var(), 3))
        print("Range: ", round(col.max() - col.min(), 3))
        print("Minimum: ", round(col.min(), 3))
        print("Maximum: ", round(col.max(), 3))
        print("25th Percentile (Q1): ", round(col.quantile(0.25), 3))
        print("75th Percentile (Q3): ", round(col.quantile(0.75), 3))
        print("Interquartile Range (IQR): ", round(col.quantile(0.75) - col.quantile(0.25), 3))
        print("Skewness: ", round(col.skew(), 3))
        print("Kurtosis: ", round(col.kurtosis(), 3))
    else:
        # Data non-numerik (string)
        print("Unique Values: ", col.unique())
        print("Value Counts: ")
        print(col.value_counts(normalize=True).round(3))

    print("\n" + "-"*50 + "\n")
```

```python
'''
Outliner merupakan data yang berbeda secara signifikan dengan data data lainnya
pada data health.csv terdapat outlier pada column numerik selain age
outlier dapat ditangani dengan beberapa metode seperti removing outliners, Quantile based capping,

1. Menghapus Outlier:

    Cara yang paling sederhana adalah dengan menghapus outlier dari dataset. Namun, perlu dipastikan bahwa outlier tersebut benar-benar merupakan kesalahan atau data yang tidak relevan. Menghapus outlier dapat dilakukan dengan menggunakan kriteria tertentu, seperti z-score atau IQR.

2. Transformasi Data:

    Transformasi data seperti logaritma, akar kuadrat, atau transformasi lainnya dapat digunakan untuk mengurangi dampak outlier. Metode ini berguna untuk membuat distribusi data lebih simetris dan mendekati distribusi normal.

3. Winsorizing:

    Winsorizing adalah metode yang mengganti nilai outlier dengan nilai yang terletak pada batas atas atau batas bawah tertentu. Misalnya, jika Anda memutuskan untuk melakukan winsorizing pada kuartil 5% dan 95%, nilai outlier yang lebih kecil dari kuantil 5% akan diganti dengan nilai kuantil 5%, sedangkan nilai outlier yang lebih besar dari kuantil 95% akan diganti dengan nilai kuantil 95%.

4. Binning atau Capping:

    Binning atau capping adalah metode di mana nilai-nilai yang ekstrem dikelompokkan ke dalam bin tertentu atau ditetapkan batas atas dan batas bawah tertentu. Ini dapat dilakukan untuk mengurangi variasi ekstrem dalam data.

kami menggunakan metode menghapus outliner 
terdapat outlier pada column numerik selain age, outlier dihandle dengan metode IQR'''


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data dari CSV
df = pd.read_csv("health.csv")

# Menghilangkan kolom 'Unnamed: 0' jika ada
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Fungsi untuk mendeteksi outlier menggunakan IQR
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

# Fungsi untuk menangani outlier dengan mengganti nilai di luar bound dengan nilai bound
def handle_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return np.where(data < lower_bound, lower_bound, np.where(data > upper_bound, upper_bound, data))

# Menangani outlier pada semua kolom numerik menggunakan metode IQR
numeric_columns = df.select_dtypes(include=[np.number]).columns
for column in numeric_columns:
    print(f"Processing column: {column}")
    outliers = detect_outliers_iqr(df[column])
    if not outliers.empty:
        print(f"Outliers detected in column {column}:")
        print(outliers)
        print()

        # Visualisasi sebelum penanganan outlier
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[column])
        plt.title(f"Boxplot of {column} before handling outliers")
        plt.show()

        # Menangani outlier
        df[column] = handle_outliers_iqr(df[column])

        # Visualisasi setelah penanganan outlier
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[column])
        plt.title(f"Boxplot of {column} after handling outliers")
        plt.show()
    else:
        print(f"No outliers detected in column {column}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari CSV
df = pd.read_csv("health.csv")

# Menghilangkan kolom 'Unnamed: 0' jika ada
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Menampilkan histogram dan boxplot untuk kolom numerik
numeric_columns = df.select_dtypes(include=[np.number]).columns
for column in numeric_columns:
    plt.figure(figsize=(12, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.xlabel(column)

    plt.tight_layout()
    plt.show()

# Menampilkan histogram untuk kolom string
string_columns = df.select_dtypes(include=[object]).columns
for column in string_columns:
    plt.figure(figsize=(8, 6))

    # Histogram
    sns.histplot(df[column], discrete=True)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Membaca data dari CSV
df = pd.read_csv("health.csv")

# Menghilangkan kolom 'Unnamed: 0' jika ada
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Mengonversi nilai infinity (inf) ke NaN (Not a Number)
df.replace([np.inf, -np.inf], np.nan, inplace=True)

numeric_columns = df.select_dtypes(include=[np.number]).columns
for column in numeric_columns:
    plt.figure(figsize=(12, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")

    # Q-Q plot
    plt.subplot(1, 2, 2)
    stats.probplot(df[column].dropna(), dist="norm", plot=plt)
    plt.title(f"Q-Q Plot of {column}")

    plt.tight_layout()
    plt.show()

print("dari hasil visualisasi semua kolom  terdistribusi normal kecuali kolom `height` yang terdistribusi skewed kanan dan kolom `age` yang terdistribusi hampir uniform")
```
