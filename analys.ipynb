```python
'''
Outliner merupakan data yang berbeda secara signifikan dengan data data lainnya
pada data health.csv terdapat outlier pada column numerik selain age
outlier dapat ditangani dengan beberapa metode seperti removing outliners, Quantile based capping,

1. Menghapus Outlier:

    Cara yang paling sederhana adalah dengan menghapus outlier dari dataset. Namun, perlu dipastikan bahwa outlier tersebut benar-benar merupakan kesalahan atau data yang tidak relevan. Menghapus outlier dapat dilakukan dengan menggunakan kriteria tertentu, seperti z-score atau IQR.

2. Transformasi Data:

    Transformasi data seperti logaritma, akar kuadrat, atau transformasi lainnya dapat digunakan untuk mengurangi dampak outlier. Metode ini berguna untuk membuat distribusi data lebih simetris dan mendekati distribusi normal.

3. Winsorizing:

    Winsorizing adalah metode yang mengganti nilai outlier dengan nilai yang terletak pada batas atas atau batas bawah tertentu. Misalnya, jika Anda memutuskan untuk melakukan winsorizing pada kuartil 5% dan 95%, nilai outlier yang lebih kecil dari kuantil 5% akan diganti dengan nilai kuantil 5%, sedangkan nilai outlier yang lebih besar dari kuantil 95% akan diganti dengan nilai kuantil 95%.

4. Binning atau Capping:

    Binning atau capping adalah metode di mana nilai-nilai yang ekstrem dikelompokkan ke dalam bin tertentu atau ditetapkan batas atas dan batas bawah tertentu. Ini dapat dilakukan untuk mengurangi variasi ekstrem dalam data.

kami menggunakan metode menghapus outliner 
terdapat outlier pada column numerik selain age, outlier dihandle dengan metode IQR'''


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data dari CSV
df = pd.read_csv("health.csv")

# Menghilangkan kolom 'Unnamed: 0' jika ada
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Fungsi untuk mendeteksi outlier menggunakan IQR
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

# Fungsi untuk menangani outlier dengan mengganti nilai di luar bound dengan nilai bound
def handle_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return np.where(data < lower_bound, lower_bound, np.where(data > upper_bound, upper_bound, data))

# Menangani outlier pada semua kolom numerik menggunakan metode IQR
numeric_columns = df.select_dtypes(include=[np.number]).columns
df = df[df['Exercise_Hours_Per_Week'] >= 0]
for column in numeric_columns:
    print(f"Processing column: {column}")
    outliers = detect_outliers_iqr(df[column])
    if not outliers.empty:
        print(f"Outliers detected in column {column}:")
        print(outliers)
        print()

        # Visualisasi sebelum penanganan outlier
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[column])
        plt.title(f"Boxplot of {column} before handling outliers")
        plt.show()

        # Menangani outlier
        df[column] = handle_outliers_iqr(df[column])

        # Visualisasi setelah penanganan outlier
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[column])
        plt.title(f"Boxplot of {column} after handling outliers")
        plt.show()
    else:
        print(f"No outliers detected in column {column}.")
```

```python
import math

def mean(data):
    return sum(data) / len(data)

def variance(data, ddof=0):
    mean_value = mean(data)
    return sum((x - mean_value) ** 2 for x in data) / (len(data) - ddof)

def standard_deviation(data, ddof=0):
    return math.sqrt(variance(data, ddof))

def t_statistic(sample_data, population_mean):
    sample_mean = mean(sample_data)
    sample_std_dev = standard_deviation(sample_data, ddof=1)
    n = len(sample_data)
    return (sample_mean - population_mean) / (sample_std_dev / math.sqrt(n))

def p_value_t(t_stat, dk, alternative='two-sided'):
    # Using the cumulative distribution function (CDF) of the t-distribution
    def t_cdf(t, dk):
        from scipy.special import betainc
        x = dk / (t**2 + dk)
        return 0.5 * betainc(0.5 * dk, 0.5, x)

    if alternative == 'less':
        return 1 - t_cdf(t_stat, dk)
    elif alternative == 'greater':
        return t_cdf(t_stat, dk)
    else:  # two-sided
        if t_stat > 0:
            return 2 * (1 - t_cdf(t_stat, dk))
        else:
            return 2 * t_cdf(t_stat, dk)

def one_sample_t_test(data, population_mean, alpha=0.05, alternative='two-sided'):
    t_stat = t_statistic(data, population_mean)
    dk = len(data) - 1
    p_val = p_value_t(t_stat, dk, alternative)
    
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_val}")
    
    if p_val < alpha:
        print("Tolak H0")
    else:
        print("Gagal Menolak H0")

def z_statistic_proportion(sample_success, sample_size, population_proportion):
    sample_proportion = sample_success / sample_size
    return (sample_proportion - population_proportion) / math.sqrt((population_proportion * (1 - population_proportion)) / sample_size)

def p_value_z(z_stat, alternative='two-sided'):
    def normal_cdf(x):
        return (1.0 + math.erf(x / math.sqrt(2.0))) / 2.0

    if alternative == 'greater':
        return 1 - normal_cdf(z_stat)
    elif alternative == 'less':
        return normal_cdf(z_stat)
    else:  # two-sided
        if z_stat > 0:
            return 2 * (1 - normal_cdf(z_stat))
        else:
            return 2 * normal_cdf(z_stat)

def proportion_test(sample_success, sample_size, population_proportion, alpha=0.05, alternative='two-sided'):
    z_stat = z_statistic_proportion(sample_success, sample_size, population_proportion)
    p_val = p_value_z(z_stat, alternative)
    
    
    
    if p_val < alpha:
        print("Tolak H0")
    else:
        print("Gagal Menolak H0")

mean_weight = mean(df["Weight"])

print("1")
print(f"Mean Weight: {mean_weight}")

one_sample_t_test(df["Weight"],65,0.05, alternative='greater')


```

```python
import math
from scipy.stats import t, f, norm

# Fungsi untuk menghitung mean
def mean(data):
    return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def variance(data, ddof=0):
    mean_value = mean(data)
    return sum((x - mean_value) ** 2 for x in data) / (len(data) - ddof)

# Fungsi untuk menghitung standar deviasi
def standard_deviation(data, ddof=0):
    return math.sqrt(variance(data, ddof))

# Fungsi untuk menghitung t-statistik untuk uji dua sampel independen
def t_statistic_two_sample(sample1, sample2):
    mean1 = mean(sample1)
    mean2 = mean(sample2)
    var1 = variance(sample1, ddof=1)
    var2 = variance(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)
    
    pooled_std_dev = math.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    t_stat = (mean1 - mean2) / (pooled_std_dev * math.sqrt(1/n1 + 1/n2))
    return t_stat

# Fungsi untuk menghitung p-value dari t-statistik menggunakan scipy
def p_value_t(t_stat, df, alternative='two-sided'):
    if alternative == 'greater':
        return 1 - t.cdf(t_stat, df)
    elif alternative == 'less':
        return t.cdf(t_stat, df)
    else:  # two-sided
        return 2 * (1 - t.cdf(abs(t_stat), df))


# Membagi data menjadi dua bagian



# Uji Hipotesis 1: Rata-rata berat badan
weight_early = early_data['Weight'].dropna().tolist()
weight_late = late_data['Weight'].dropna().tolist()
t_stat_weight = t_statistic_two_sample(weight_early, weight_late)
df_weight = len(weight_early) + len(weight_late) - 2
p_value_weight = p_value_t(t_stat_weight, df_weight)

print(f"T-Statistic for Weight: {t_stat_weight}, P-Value: {p_value_weight}")
if p_value_weight < 0.05:
    print("Tolak H0: Rata-rata berat badan tidak sama.")
else:
    print("Gagal Menolak H0: Rata-rata berat badan sama.")

# Verifikasi dengan scipy ttest_ind
from scipy.stats import ttest_ind

t_stat_scipy, p_value_scipy = ttest_ind(weight_early, weight_late)
print(f"Scipy T-Statistic: {t_stat_scipy}, Scipy P-Value: {p_value_scipy}")
```

```python
# Fungsi untuk uji F dua sampel (perbandingan variansi)
def f_statistic_two_sample(sample1, sample2):
    var1 = variance(sample1, ddof=1)
    var2 = variance(sample2, ddof=1)
    return var1 / var2

# Uji Hipotesis 3: Variansi Tekanan Darah
bp_early = early_data['Blood_Pressure'].dropna().tolist()
bp_late = late_data['Blood_Pressure'].dropna().tolist()
f_stat_bp = f_statistic_two_sample(bp_early, bp_late)
df1_bp = len(bp_early) - 1
df2_bp = len(bp_late) - 1
p_value_bp = 2 * min(f.cdf(f_stat_bp, df1_bp, df2_bp), 1 - f.cdf(f_stat_bp, df1_bp, df2_bp))

print(f"F-Statistic for Blood Pressure: {f_stat_bp}, P-Value: {p_value_bp}")
if p_value_bp < 0.05:
    print("Tolak H0: Variansi tekanan darah tidak sama.")
else:
    print("Gagal Menolak H0: Variansi tekanan darah sama.")

# Verifikasi dengan scipy bartlett
from scipy.stats import bartlett

t_stat_scipy, p_value_scipy = bartlett(bp_early, bp_late)
print(f"Scipy t-Statistic for Blood Pressure: {t_stat_scipy}, Scipy P-Value: {p_value_scipy}")
```
